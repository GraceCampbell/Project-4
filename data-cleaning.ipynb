# Data Cleaning

```python
import pandas as pd
```

```python
housing_assistance = pd.read_csv('materials/housing_assistance.csv')
```

```python
housing_assistance.shape
```

```python
housing_assistance.head()
```

```python
housing_assistance.columns = ['disaster', 'state', 'county', 'city', 'zip', 'valid_registrations', 'avg_inspected_damage',
                             'total_inspected', 'total_damage', 'no_inspected_damage', 'damage_1_10000', 'damage_10000_20000',
                             'damage_20000_30000', 'damage_30000_max', 'approved_for_assistance', 'total_approved_amt',
                             'repair_replace_amt', 'rental_amt', 'other_needs_amt', 'approved_1_10000', 'approved_10001_25000',
                             'approved_25001_max', 'total_max_grants']
```

```python
for col in housing_assistance.columns:
    housing_assistance[col] = housing_assistance[col].apply(lambda x: str(x).strip('$'))
    housing_assistance[col] = housing_assistance[col].apply(lambda x: str(x).replace(',', ''))
    housing_assistance[col] = housing_assistance[col].apply(lambda x: int(x) if x.isdigit() else x)
```

```python
housing_assistance.dtypes.value_counts()
```

```python
housing_assistance.head()
```

```python
disasters = pd.read_csv('materials/filtered_incident_type.csv')

disasters = disasters[['Disaster Number', 'Incident Type']]
disasters.head()
```

```python
disaster_codes = (set(zip(disasters['Disaster Number'].values, disasters['Incident Type'].values)))
```

```python
codes_dict = {num: dis for num, dis in disaster_codes}
```

```python
housing_assistance['disaster_type'] = housing_assistance['disaster'].map(codes_dict)
```

```python
housing_assistance.dropna(inplace=)
```

```python
housing_assistance.to_csv('materials/housing_cleaned.csv', index=False)
```
